```python
import numpy as np
from goatools.go_enrichment import GOEnrichmentStudy
from goatools import obo_parser
import wget
import os, pandas as pd
```

```python
#THIS IS A WAY TO PARSE THE FULL GO TREE
go_obo_url = 'http://purl.obolibrary.org/obo/go/go-basic.obo'
data_folder = '/cluster/shz24/data/go'

# Check if we have the ./data directory already
if(not os.path.isfile(data_folder)):
    # Emulate mkdir -p (no error if folder exists)
    try:
        os.makedirs(data_folder)
    except OSError as e:
        if(e.errno != 17):
            raise e
else:
    raise Exception('Data path (' + data_folder + ') exists as a file. '
                   'Please rename, remove or change the desired location of the data path.')

# Check if the file exists already
if(not os.path.isfile(data_folder+'/go-basic.obo')):
    go_obo = wget.download(go_obo_url, data_folder+'/go-basic.obo')
else:
    go_obo = data_folder+'/go-basic.obo'

# Import the OBO parser from GOATools
go = obo_parser.GODag(go_obo)
```

```python
def embry_num(name):
    all_assoc_names = pd.concat([
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:go[x.GO_ID].name,axis=1),
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:", ".join([e.name for e in go[x.GO_ID].children]),axis=1).rename("child_names"), 
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:", ".join([e.name for e in go[x.GO_ID].parents]),axis=1).rename("parent_names")
    ],axis = 1)
    allwords = pd.Series([e for l in all_assoc_names.melt().value for e in l.split(" ")]).value_counts().rename("counts").reset_index()
    if allwords.empty:
        return 0
    return allwords.loc[allwords["index"].str.contains("embry")].counts.sum()
```

```python
def neuro_num(name):
    all_assoc_names = pd.concat([
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:go[x.GO_ID].name,axis=1),
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:", ".join([e.name for e in go[x.GO_ID].children]),axis=1).rename("child_names"), 
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:", ".join([e.name for e in go[x.GO_ID].parents]),axis=1).rename("parent_names")
    ],axis = 1)
    allwords = pd.Series([e for l in all_assoc_names.melt().value for e in l.split(" ")]).value_counts().rename("counts").reset_index()
    if allwords.empty:
        return 0
    return allwords.loc[allwords["index"].str.contains("neur")].counts.sum()
```

```python
def heart_num(name):
    all_assoc_names = pd.concat([
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:go[x.GO_ID].name,axis=1),
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:", ".join([e.name for e in go[x.GO_ID].children]),axis=1).rename("child_names"), 
        ontologies.loc[ontologies.DB_Object_Symbol==name].apply(lambda x:", ".join([e.name for e in go[x.GO_ID].parents]),axis=1).rename("parent_names")
    ],axis = 1)
    allwords = pd.Series([e for l in all_assoc_names.melt().value for e in l.split(" ")]).value_counts().rename("counts").reset_index()
    if allwords.empty:
        return 0
    return allwords.loc[allwords["index"].str.contains("heart")].counts.sum()
```
